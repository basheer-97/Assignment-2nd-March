```
data1.createOrReplaceTempView("customer")
data2.createOrReplaceTempView("fd_account")
data3.createOrReplaceTempView("rd_account")
data4.createOrReplaceTempView("sb_account")
```

```
#1
%sql
select d1.CUST_FNAME,d2.ACCOUNT_NO,d2.CUST_ID,d2.INITIAL_AMT from customer as d1 join fd_account as d2
on d1.CUST_ID=d2.CUST_ID
where INITIAL_AMT>25000 AND ACC_STATUS='Active'
order by INITIAL_AMT desc; 
```

```
#2
%sql
select c.CUST_ID,c.CUST_FNAME,fd.ACCOUNT_NO,fd.INITIAL_AMT,rd.CURR_BALANCE,sb.CURR_BAL_AMT,(fd.INITIAL_AMT+rd.CURR_BALANCE+sb.CURR_BAL_AMT)as networth from fd_account as fd join rd_account as rd on fd.CUST_ID=rd.RD_CUST_ID
join sb_account as sb on rd.RD_CUST_ID=sb.PRIMARY_CUST_ID
join customer as c on c.CUST_ID=rd.RD_CUST_ID
where fd.ACC_STATUS='Active' and rd.ACC_STATUS='Active' and sb.ACC_STATUS='Active'
order by networth desc
```

```
#4
%sql
SELECT fd_account.CUST_ID, fd_account.ACCOUNT_NO,fd_account.INITIAL_AMT AS AMT, fd_account.ACC_STATUS FROM fd_account WHERE fd_account.ACC_STATUS='Active'
UNION
SELECT rd_account.RD_CUST_ID, rd_account.ACCOUNT_NO, rd_account.CURR_BALANCE, rd_account.ACC_STATUS FROM rd_account WHERE rd_account.ACC_STATUS = 'Active'
UNION
SELECT sb_account.PRIMARY_CUST_ID, sb_account.ACCOUNT_NO, sb_account.CURR_BAL_AMT, sb_account.ACC_STATUS FROM sb_account WHERE sb_account.ACC_STATUS = 'Active'
ORDER BY CUST_ID, ACCOUNT_NO
```

```
#5
%sql
(select c.cust_id as CustomerID,f.account_no as AccountNo,f.initial_amt as BalanceAmmount,f.acc_status as AccountStatus,'FD' as acc_type
from customer c 
left join fd_account f on c.cust_id=f.cust_id
where f.acc_status=='Active')
                              union
                              
(select c.cust_id,s.account_no,s.curr_bal_amt,s.acc_status,'SB'
from customer c 
left join sb_account s on c.cust_id=s.primary_cust_id
where s.acc_status=='Active')                                       
                             union                           
(select c.cust_id,r.account_no,r.curr_balance,r.acc_status,'RD'
from customer c 
left join rd_account r on c.cust_id=r.rd_cust_id
where r.acc_status=='Active')
order by CustomerID,AccountNo;
```
